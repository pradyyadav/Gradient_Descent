# Gradient Descent

## Importing Libraries

```python
import numpy as np 
import matplotlib.pyplot as plt
```

## Creating random data

```python
from sklearn.datasets.samples_generator import make_regression
X,y = make_regression(n_samples = 500,n_features = 1,n_informative=1,noise=6,bias=10,random_state = 200)
```

```python
y[:5]
```

## Plotting data

```python
plt.scatter(X,y)
```

## Gradient Descent Implementation

```python
def gradient_descent(X,y):
    m = b = 0
    learning_rate = 0.01      # It is the small steps by which the cost function is optimised along the slope
    epochs = 1000             # The iterations during which cost function is optimised
    n = len(X)
    for epoch in range(epochs):
        y_predicted = m*X + b                           # Predicted function with respect to features 
        cost = (1/(2*n))*sum([val**2 for val in (y_predicted-y)])  # This is the cost function
        derivative_wrt_m = (1/n)*sum((y_predicted-y)*X)         # Gives the slope of cost function along m
        derivative_wrt_b = (1/n)*sum(y_predicted-y)             # Gives the slope of cost function along b 
        m = m - learning_rate * derivative_wrt_m                   # Steps by which cost is minimized along m
        b = b - learning_rate * derivative_wrt_b                   # Steps by which cost is minimized along b 
        print("m: {},  b: {},  Epoch: {},  cost: {}   ".format(m,b,epoch,cost))
    return m,b,cost
```

## Setting feature and target variables

```python
features = np.array(X).reshape(-1)
target = np.array(y)
```

## Fitting the data

```python
slope,intercept,cost = gradient_descent(features,target)
```

### When cost : 127.567, At epoch : 6

```python
slope_cost0 = 0.82765
inter_cost0 = 0.70078
plt.scatter(X,y,color='y')
plt.plot(X,slope_cost0*X+inter_cost0,linewidth = 3)
```

### When cost : 107.061, At epoch : 16

```python
slope_cost1 = 1.91262
inter_cost1 = 1.61750
plt.scatter(X,y,color='y')
plt.plot(X,slope_cost1*X+inter_cost1,linewidth = 3)
```

### When cost : 50.604, At epoch : 64

```python
slope_cost2 = 5.83075
inter_cost2 = 4.90526
plt.scatter(X,y,color='y')
plt.plot(X,slope_cost2*X+inter_cost2,linewidth = 3)
```

### When cost : 33.000, At epoch : 100

```python
slope_cost3 = 7.74237
inter_cost3 = 6.49164
plt.scatter(X,y,color='y')
plt.plot(X,slope_cost3*X+inter_cost3,linewidth = 3)
```

## Optimal values of slope and intercept

```python
plt.scatter(X,y,color='c')
plt.plot(X,slope*X+intercept,color='r',linewidth = 3)
```

Hence, gradient descent runs iteratively (as given in epochs) and gives the optimal parameters corresponding to the minimum cost.
